```python
#loading the data

#importing pandas
import pandas as pd

#loading the files
file1= pd.read_csv("online_retail_2009.csv")
file2= pd.read_csv("online_retail.csv")

#combining the files
retail_data = pd.concat([file1, file2], ignore_index=True)
retail_data.info()
```

```python
#previewing the data
retail_data.head(15)
```

The StockCode has five numbers or five numbers and a letter. 

Finding out if there are other alternatives

```python
#filtering data with five numbers only using regular expressions with the 'str.match() method provided by pandas

five_numbers = retail_data[retail_data['StockCode'].str.match(r'^\d{5}$')]
five_numbers.info()
```

```python
#filtering data with numbers and letter  using regular expressions with the 'str.match() method provided by pandas
num_letters = retail_data[retail_data['StockCode'].str.match('^\d{5}[a-zA-Z]{1,2}$')]
num_letters.info()
```

```python
num_letters.head(10)
```

When the Stockcode has a letter(s) at the end, it a variant of a product. 

```python
#finding if there are more variants  using regular expressions with the 'str.match() method provided by pandas
other_formats = retail_data[~retail_data['StockCode'].str.match('^\d{5}$|^\d{5}[a-zA-Z]{1,2}$')]
other_formats.info()
```

```python
other_formats.head(15)
```

```python
#evaluating the number of unique items per column

print(other_formats.nunique())
```

```python
#finding the missing values per column 

print(other_formats.isnull().sum())
```

```python
#print out any unique StockCode and Description Combination in the other 

#droping rowss based on the two columns 
other_formats_clean = other_formats.dropna(subset=['StockCode', 'Description'])
other_formats_codes = other_formats_clean.drop_duplicates(subset=['StockCode','Description'])

print(other_formats_codes)
```

```python
# I want to know the number of Unique values in the other formats 

#grouping the data
stock_counts =  other_formats.groupby(['StockCode', 'Description']).size().reset_index(name='Count')

#sorting the data in descending order
sorted_stock_counts = stock_counts.sort_values(by = 'Count', ascending = False)

print(sorted_stock_counts)
```

Removing all the rows below samples, because there is nobody to explain why the format is different. 

I will drop them because I suppose that this were old products, and considering they are few observations, they will not have such an impact. 

However, it would have been useful for the stalkholder to expound why the values are different. 



```python
retail_data_filtered = retail_data[retail_data['StockCode'].str.match('^\d{5}$|^\d{5}[a-zA-Z]{1,2}$') | 
                                   retail_data['StockCode'].isin(['POST', 'DOT', 'M', 'C2','D'])]

retail_data_filtered.info()
```

Reduced the number of observations from by 646 from 1067371 to 1066725

```python
#Evaluating the number of missing values 
retail_data_filtered.isnull().sum()
```

We have StockCodes, but we have missing values on the Description Column. Therefore, I would like us to fill the missing values(Description column) by scanning the Stockcodes. 

```python
#creating a copy of the dataframe
retail_copy = retail_data_filtered.copy()

# Grouping the data for each StockCode (creates a sub dataframe)
# The transform method applies a particular transformation to each description column in each group
#x.ffill() forward fills and bfill() backward fills the NaN values in the Description column 
retail_copy['Description'] = retail_copy.groupby('StockCode')['Description'].transform(lambda x: x.ffill().bfill())

retail_data_filtered = retail_copy

retail_data_filtered.isnull().sum()
```

This has reduced the number of missing values(Description column) from 4331 to 341




```python
missing_descriptions = retail_data_filtered[retail_data_filtered['Description'].isnull()]
#missing_descriptions.head(50)
missing_descriptions.tail(20)
```

Investigating this further, there is a trend of having zero values in the price column and negative values for the quantity column. 

Therefore, let's evaluate the number of unique values for each column

```python
#evaluating unique values 
missing_descriptions.nunique()
```

Having confirmed that the rows missing a Description also have 0.0 in the price. 

The proposal would have to seek explanations on the same, but since that is not feasible, the next best action is to drop this values. 

This will reduce the number of observations to 1066384

```python
retail_data_filtered= retail_data_filtered.dropna(subset=['Description'])

retail_data_filtered.nunique()
```

Before evaluating the price point, there are more descriptions than existing StockCode. We evaluate the StockCodes with more than one description 



```python
#group the data using StockCodes, and declare the level of aggregation as unique description 

grouped = retail_data_filtered.groupby('StockCode').agg({'Description':'nunique'})

#filter out StockCodes with more than one description
multiple_descriptions = grouped[grouped['Description']>1].index

#creating a subset of the data 
subset = retail_data_filtered[retail_data_filtered['StockCode'].isin(multiple_descriptions)]

#dropping duplicates based on StockCode and Description 

subset = subset.drop_duplicates(subset=['StockCode', 'Description'])

subset = subset.sort_values(by='StockCode')

subset.head(15)
```

A very intresting development, some of the StockCodes have repeated descriptions showing negative changes to the stock, be it returns, damages, etc. While there are spelling errors or different products(Description) that share the same StockCode

Separate the dataset into two, one with the negative QTY values and one with positive values.
