```python
#loading the data

#importing pandas
import pandas as pd

#loading the files
file1= pd.read_csv("online_retail_2009.csv")
file2= pd.read_csv("online_retail.csv")

#combining the files
retail_data = pd.concat([file1, file2], ignore_index=True)
retail_data.info()
```

```python
#previewing the data
retail_data.head(15)
```

The StockCode has five numbers or five numbers and a letter. 

Finding out if there are other alternatives

```python
#filtering data with five numbers only 

five_numbers = retail_data[retail_data['StockCode'].str.match(r'^\d{5}$')]
five_numbers.info()
```

```python
#filtering data with numbers and letter
num_letters = retail_data[retail_data['StockCode'].str.match('^\d{5}[a-zA-Z]$')]
num_letters.info()
```

```python
num_letters.head(10)
```

When the Stockcode has a letter at the end, it a variant of a product.
